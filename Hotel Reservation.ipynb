```python
df_hotel.drop(["Booking_ID"], axis = 1, inplace = True)
```

```python
colname=[]
for x in df_hotel.columns:
    if df_hotel[x].dtype=="object":
        colname.append(x)
colname
```

```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()

for x in colname:
    df_hotel[x]=le.fit_transform(df_hotel[x])
    
    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    print('Feature', x)
    print('mapping', le_name_mapping)
```

```python
X=df_hotel.values[:,0:-1]
Y=df_hotel.values[:,-1]
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X=scaler.fit_transform(X)#fit will lear mean and sigma values 
#scaler.fit_transform(X)
#X=scaler.transform(X)
print(X)
```

```python
#SPLITTING THE DATA
from sklearn.model_selection import train_test_split

#split the data into test and train 
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2,random_state=10)
```

# Building model logistic regression

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
# BUILDING MODEL
from sklearn.linear_model import LogisticRegression
#create a model object 
classifier=LogisticRegression()
#train the model object
classifier.fit(X_train,Y_train)

Y_pred=classifier.predict(X_test)
print(Y_pred)
```

```python
#print(list(zip(Y_test,Y_pred)))

print(list(zip(df_hotel.columns[:-1], classifier.coef_.ravel()))) # coef_.ravel is used to flatten the array to avoid the problem of 1D and 2D Clash 


print(classifier.intercept_)
#classifier.coef_
```

```python
#Evalution matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

# KNN

```python
X=df_hotel.values[:,0:-1]
Y=df_hotel.values[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler.fit(X)#scaler object will learn on mu and sigma of all the variables
#scaler.fit_transform(X_train)
X=scaler.transform(X)#the knowledge from scaler object will be used to calculate the z score 
```

```python
#SPLITTING THE DATA
from sklearn.model_selection import train_test_split

#split the data into test and train 
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2,random_state=10)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
# predicting using the Kneighbors Classifier
from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=int(np.sqrt(len(X_train))), metric='minkowski')

#euclidean, manhatten, minkowski
#fit the model on the data and predict the values
model_KNN.fit(X_train, Y_train)

Y_pred=model_KNN.predict(X_test)
#print(list(zip(Y_test, Y_pred)))
```

```python
#Evalution matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

# Hperparameter-KNN

```python
np.sqrt(len(X_train))
```

```python
from sklearn.metrics import accuracy_score
my_dict={}
for K in range(1,21):
    model_KNN = KNeighborsClassifier(n_neighbors=K,metric="euclidean")
    model_KNN.fit(X_train, Y_train) 
    Y_pred = model_KNN.predict(X_test)
    print ("Accuracy is ", accuracy_score(Y_test,Y_pred), "for K-Value:",K)
    my_dict[K]=accuracy_score(Y_test,Y_pred)
```

```python
for k in my_dict:
    if my_dict[k]==max(my_dict.values()):
        print(k,":",my_dict[k])
```

```python
# predicting using the Kneighbors Classifier
from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=7, metric='manhattan')

#euclidean, manhatten, minkowski
#fit the model on the data and predict the values
model_KNN.fit(X_train, Y_train)

Y_pred=model_KNN.predict(X_test)
#print(list(zip(Y_test, Y_pred)))
```

```python
#Evalution matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```
